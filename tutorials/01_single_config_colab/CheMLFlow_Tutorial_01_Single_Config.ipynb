## 5. Read the output metrics

```python
from pathlib import Path
import json

run_dir = Path("tutorials/01_single_config_colab/artifacts/runs/pgp_svm_cv_fold0")
metrics = json.loads((run_dir / "svm_metrics.json").read_text())
split_metrics = json.loads((run_dir / "svm_split_metrics.json").read_text())

print("Run dir:", run_dir)
print("\nPrimary metrics:")
print(json.dumps(metrics, indent=2))
print("\nSplit metrics:")
print(json.dumps(split_metrics, indent=2))
```
